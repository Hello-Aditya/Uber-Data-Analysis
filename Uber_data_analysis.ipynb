```python
# which month is having more round trips
#same_month=pd.DatetimeIndex(uber[uber['START']==uber['STOP']]['START_DATE']).month
round_trip=uber[uber['START']==uber['STOP']]['month']
#same_month
round_trip
sns.countplot(round_trip,order=round_trip.value_counts().index)
```

```python
#visualize the ratio how of much round trip booked respective of total bookings
def round(x):
    if x['START']==x['STOP']:
        return 'yes'
    else:
        return 'no'
```

```python
uber['Round_trip']=uber.apply(round,axis=1)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(uber['Round_trip'],order=uber.Round_trip.value_counts().index)
```

```python
#method_2 for above question
plt.figure(figsize=(12,7))
sns.countplot(uber['Round_trip'],hue=uber['month'],palette='viridis')
plt.legend()
```
